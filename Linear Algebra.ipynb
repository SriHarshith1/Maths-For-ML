**The Dot Product**

If we have two vectors (say, ***x*** and ***y***) with the same length *n*, we can calculate the dot product between them. This is annotated several different ways, including the following:

* $x \cdot y$
* $x^Ty$
* $\langle x,y \rangle$

Regardless which notation you use (I prefer the first), the calculation is the same; we calculate products in an element-wise fashion and then sum reductively across the products to a scalar value. That is, $x \cdot y = \sum_{i=1}^{n} x_i y_i$

The dot product is ubiquitous in deep learning: It is performed at every artificial neuron in a deep neural network, which may be made up of millions (or orders of magnitude more) of these neurons

# Solving Linear Equations

In the **Substitution** example, the two equations in the system are:
$$ y = 3x $$
$$ -5x + 2y = 2 $$

The second equation can be rearranged to isolate $y$:
$$ 2y = 2 + 5x $$
$$ y = \frac{2 + 5x}{2} = 1 + \frac{5x}{2} $$

```python
x = np.linspace(-10, 10, 1000)
```

```python
y1 = 3 * x
```

```python
y2 = 1 + (5*x)/2
```

```python
fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')
ax.set_xlim([0, 3])
ax.set_ylim([0, 8])
ax.plot(x, y1, c='green')
ax.plot(x, y2, c='blue')
plt.axvline(x=2, color='purple', linestyle='--')
_ = plt.axhline(y=6, color='purple', linestyle='--')
```

In the **Elimination** example, the two equations in the system are:
$$ 2x - 3y = 15 $$
$$ 4x + 10y = 14 $$

Both equations can be rearranged to isolate $y$. Starting with the first equation:
$$ -3y = 15 - 2x $$
$$ y = \frac{15 - 2x}{-3} = -5 + \frac{2x}{3} $$

Then for the second equation:
$$ 4x + 10y = 14 $$
$$ 2x + 5y = 7 $$
$$ 5y = 7 - 2x $$
$$ y = \frac{7 - 2x}{5} $$

```python
x = np.linspace(-10, 10, 1000)
```

```python
y1 = (2*x)/3 -5
```

```python
y2 = 1.4 - (2*x)/5
```

```python
fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')
ax.plot(x, y1, color = 'green')
ax.plot(x, y2, color = 'blue')
plt.xlim([-2, 8])
plt.ylim([-5, 5])
plt.axvline(x = 6, color = 'purple', linestyle = '--')
plt.axhline(y = -1 , color = 'purple', linestyle = '--')
plt.show()
```

To demonstrate that a matrix is orthogonal, we need to verify that all of its columns and rows are orthogonal vectors. This can be achieved by following these steps:

1. **Dot Product**: Calculate the dot product between any two distinct columns (or rows). For the matrix to be orthogonal, this dot product must equal zero.

2. **Norm**: Ensure that the norm (length) of each column (or row) is equal to 1.

If both conditions are satisfied, we can conclude that the matrix is orthogonal.

* To verify this we can also multiply k, $k^T$ if it gives an Identity matrix then the matrix is indeed orthogonal

# Affine Transformations via Matrix Application

let's say we have a vector $v$

```python
import matplotlib.pyplot as plt
```

```python
def plot_vectors(vectors, colors):
    """
    Plot one or more vectors in a 2D plane, specifying a color for each. 

    Arguments
    ---------
    vectors: list of lists or of arrays
        Coordinates of the vectors to plot. For example, [[1, 3], [2, 2]] 
        contains two vectors to plot, [1, 3] and [2, 2].
    colors: list
        Colors of the vectors. For instance: ['red', 'blue'] will display the
        first vector in red and the second in blue.
        
    Example
    -------
    plot_vectors([[1, 3], [2, 2]], ['red', 'blue'])
    plt.xlim(-1, 4)
    plt.ylim(-1, 4)
    """
    plt.figure()
    plt.axvline(x=0, color='lightgray')
    plt.axhline(y=0, color='lightgray')

    for i in range(len(vectors)):
        x = np.concatenate([[0,0],vectors[i]])
        plt.quiver([x[0]], [x[1]], [x[2]], [x[3]],
                   angles='xy', scale_units='xy', scale=1, color=colors[i],)
```

```python
v = np.array([3,1])
```

```python
plot_vectors([v], ['lightblue'])
plt.xlim(-1, 5)
_ = plt.ylim(-1, 5)
```

In addition to flipping a matrix over an axis (a.k.a., reflection), other common affine transformations include:

* Scaling (changing the length of vectors)
* Shearing 
* Rotation

# Eigen Vectors

* An eigenvector is a special vector $v$ such that when it is transformed by some matrix $A$ has the same direction as $v$

* An eigenvalue Î» is a value that simply scales the eigenvector $v$ such that following equation is satisfied

    *$Av = Î»v$*




**Using NumPy**

```python
A = A.numpy()
```

```python
A
```

```python
lambdas, V = np.linalg.eig(A) 
```

```python
V  # each column is a seperate eigenvector v
```

```python
lambdas
```

```python
v = V[:,0] 
v
```

```python
lambda1 = lambdas[0]
```

```python
Av = np.dot(A,v)
```

```python
Î»v = np.dot(lambda1,v)
```

```python
Av == Î»v
```

```python
plot_vectors([v, Av], ['blue', 'lightblue'])
plt.xlim(-2, 2)
plt.ylim(-2, 2)
```

```python
v2 = V[:, 1]
```

```python
Av2 = np.dot(A, v2)
```

```python
lambda2 = lambdas[1]
```

```python
Î»v2 = np.dot(lambda2,v2)
```

```python
Î»v2, Av2
```

```python
Av2 == Î»v2
```

```python
plot_vectors([v2, Av2], ['blue', 'lightblue'])
plt.xlim(-1, 5)
plt.ylim(-3, 2)
```

* Matrix is singular if one of the eigenvalues is zero

maximum of $f(x)$ = largest eigen value



minimum of $f(x)$ = smallest eigen value

Eigen decomposition is a mathematical technique used in linear algebra to break down a square matrix into its constituent parts: eigenvalues and eigenvectors. This decomposition provides insights into the properties of the matrix and simplifies various computations.

# Singular Value Decomposition

* Unlike eigendecompostion, which is applicable to square matrices only, SVD is applicable to any real-values matrix

* Decomposes a amtrix into :

        * Singular Vectors (Analogus to eigenvectors)
        * Singular Values (Analogus to eigenvalues)

* For some matrix $A$ , it's SVD is $A = UDV^T$

Where:

* U is an Orthogonal $ m$ x  $m $ matrix; It's columns are the left-singular vectors of $A$        
* V is an Orthogonal $n$ x $n$ matrix; It's columns are the right-singular vectors of $A$
* D is an Orthogonal $m$ x $n$ matrix; It's columns are the singular values of $A$


```python
A = torch.tensor([[-1, 2], [3, -2], [5, 7]], dtype= torch.float)
A
```

```python
U, d, VT = torch.linalg.svd(A) #V is already transposed
```

```python
U
```

```python
VT
```

```python
d
```

```python
d = torch.diag(d)
```

* D must have same dimesnions as A for $UDV^T$ matrix multiplcation to be possible

```python
d = torch.concatenate((d, torch.tensor([[0, 0]])), axis= 0)
```

```python
torch.matmul(U, torch.matmul(d, VT))
```

# Image compression via SVD

```python
from PIL import Image
```

convert the image to grayscale so we don't have to deal with multiple color channels

Converting image data into numpy matrix, which doesn't impact the image data

```python
imgmat = np.array(list(imggray.getdata(band = 0)), float)
imgmat.shape = (imggray.size[1], imggray.size[0])
imgmat = np.matrix(imgmat)
_ = plt.imshow(imgmat, cmap = 'gray')
```

Calculate SVD of the image matrix

```python
U, d, VT = np.linalg.svd(imgmat)
```

```python
U.shape, d.shape, VT.shape
```

```python
d[:1]
```

```python
np.diag(d[:1])
```

```python
reconstruced_image = np.matrix(U[:, 1] * np.diag(d[:1]) * VT[1, :])
plt.imshow(reconstruced_image, cmap='gray')
plt.title('reconstruced_image')
plt.show()
```

# Moore-Penrose Psuedoinverse

* This allows us to inverse a non-square matrix, enabling us to solve linear systems.
* For some matrix \( A \), its pseudoinverse \( A^+ \) can be calculated by:

\[
$A^+ = VD^+U^T$
\]

Where:
- \( U, D, \) and \( V \) are the SVD (Singular Value Decomposition) of \( A \).
- \( D^+ \) is obtained by taking the reciprocal of all non-zero elements of \( D \) and transposing it.

```python
A 
```

```python
U, d, VT = np.linalg.svd(A)
```

To create $D^+$ for, we first invert the non-zero values of $d$

```python
D= np.diag(d)
```

Since $D$ is a diagnoal matrix $D^+$ can be inverted by finding the inverse of $D$.
$D^+$ must have the dimensions required to make the matrix multiplication possible

```python
Dinv = np.linalg.inv(D)
Dinv.shape
Dplus = np.concatenate((Dinv, np.array([[0, 0]]).T), axis = 1)
```

```python
Aplus = np.matmul(VT.T, np.matmul(Dplus, U.T))
```

```python
Aplus
```

* This can also be done by using existing method pinv

```python
np.linalg.pinv(A)
```

```python
torch.linalg.pinv(A)
```

# Regression with Pseudoinverse

$A^+$ is mega useful because non-sqaure matrices are common in Machine learning


* For example in a house price regression model the model is represented by $y = Xw$


Where y is the house prices , X is the features and W is the unknows or we can say them as models unknown parameters . would be unusual to have exactly as many cases (n) as features (m)

* Assuming $X^-1$ exists, matrix inversion can solve for $w$ : $w = X^-1 * y$ . we can solve this even if $ n!= m$

```python
x1 = [0, 1, 2, 3, 4, 5, 6, 7.]  #E.g : Dosage of drug for treating Alzheimer's disease
y = [1.86, 1.31, .62, .33, .09, -.67, -1.23, -1.37]  #E.g Patient's forgetfulness score
```

let's calculate the y intercept of the curve this allows the curve to not pass through (0, 0). The y-intercept is constant across all the points so we can set it equal to 1 across the board

```python
x0 = np.ones(8)
```

Concatenate $x0$ & $x1$

```python
X = np.concatenate((np.matrix(x0).T, np.matrix(x1).T), axis = 1)
X
```

We can calculate the weights of the $w$ using the equation $w = X^+y$

```python
w =  np.matmul(np.linalg.pinv(X), y)
w
```

# Trace Operator

Sum of the diagonal elements of a matrix: $$\sum_i A_{i,i}$$

```python
A
```

```python
torch.trace(A)
```

The trace operator has a number of useful properties that come in handy while rearranging linear algebra equations

* Tr($A$) = Tr($A^T)
* Tr(ABC) = Tr(CAB) = Tr(BCA)
* Trace operator can be used to calculate the forbenius norm of a matrix $$||A||_F = \sqrt{\mathrm{Tr}(AA^\mathrm{T})}$$


```python
torch.trace(A)  == torch.trace(A.T)
```

```python
torch.norm(A) == torch.sqrt(torch.trace(torch.matmul(A, A.T)))
```

These shows that the trace operator can be used to calculate norms of a matrix easily making the computation easier
