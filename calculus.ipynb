# Derivatives

\frac{dy}{dx} = \lim_{\Delta x \to 0} \frac{f(x + \Delta x) - f(x)}{\Delta x}
$$

* Using the delta method, we've derived the definition of differentiation from first principles. The derivative of $y$ (denoted $dy$) with respect to $x$ (denoted $dx$) can be represented as: 
$$\frac{dy}{dx} = \lim_{\Delta x \to 0} \frac{\Delta y}{\Delta x}$$



# Derivative Rules

1. **Constant Rule**: 
   - If $$ f(x) = c $$, then $$ f'(x) = 0 $$.

2. **Constant Multiple Rule**: 
   - If $$ g(x) = c \cdot f(x) $$, then $$ g'(x) = c \cdot f'(x) $$.

3. **Power Rule**: 
   - If $$ f(x) = x^n $$, then $$ f'(x) = n \cdot x^{n-1} $$.

4. **Sum and Difference Rule**: 
   - $$ (f + g)' = f' + g' $$
   - $$ (f - g)' = f' - g' $$

5. **Product Rule**: 
   - If $$ h(x) = f(x)g(x) $$, then $$ h'(x) = f'g + fg' $$.

6. **Quotient Rule**: 
   - If $$ h(x) = \frac{f(x)}{g(x)} $$, then 
   $$
   h'(x) = \frac{f'g - fg'}{g^2}
   $$

7. **Chain Rule**: 
   - If $$ h(x) = f(g(x)) $$, then 
   $$
   h'(x) = f'(g(x))g'
   $$

### Common Derivatives

- **Trigonometric Functions**:
  - $$ \frac{d}{dx}(\sin x) = \cos x $$
  - $$ \frac{d}{dx}(\cos x) = -\sin x $$
  
- **Exponential Functions**:
  - $$ \frac{d}{dx}(e^x) = e^x $$

- **Logarithmic Functions**:
  - $$ \frac{d}{dx}(\ln x) = \frac{1}{x} $$

These rules streamline the process of differentiation, making it easier to tackle calculus problems efficiently!

# Automatic Differentiation

* TensorFlow and PyTorch are the two most popular automatic differentiation libraries

let's use them to calculate $dy/dx$ at $x$ =5 where $y = x^2$

**Autodiff with PyTorch**

```python
import torch
```

```python
x = torch.tensor(5.0) # defining x as scalar
```

```python
x.requires_grad_() # contagiously track gradients through forward pass
```

```python
y = x**2 # defining function
```

```python
y.backward() # uses automatic differentiation
```

```python
x.grad # calculating the derivative at  x = 2
```

**Autodiff with TensorFlow**

```python
import tensorflow as tf
```

```python
x = tf.Variable(5.0)
```

```python
with tf.GradientTape() as t:
    t.watch(x)  # track forwardpass
    y = x**2
```

```python
t.gradient(y, x) # use autodiff
```

# Regression in PyTorch

* Using PyTorch automatic differentiation to fit a straight line through the data points

```python
import torch
import matplotlib.pyplot as plt
```

```python
x = torch.tensor([0, 1, 2, 3, 4, 5, 6, 7.]) #E.g: Dosage of drug for treating alzheimer's disease
```

The $y$ values were created using the eqaution of a line $y = mx + b$. This way, we know the model parameters to be learned are, say $m = -0.5$ and $b = 2$, Random, normally distributed noise has been added to simulate sampling error

```python
y = -0.5*x + 2 + torch.normal(mean = torch.zeros(8), std = 0.2)
```

```python
y
```

```python
fig, ax = plt.subplots()
plt.title('Clincal Trail')
plt.xlabel('Forgetfulness')
plt.ylabel('Drug Dosage')
_ = ax.scatter(x, y)
```

Initialize the slope paramater $m$ with a "random" near zero values . let's take m as 0.9

```python
m = torch.tensor([0.9]).requires_grad_()
```

and the same for value of intercept b . lets take b as 0.1

```python
b = torch.tensor([0.1]).requires_grad_()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def regression_plot(my_x, my_y, my_m, my_b):
    # Create a new figure and axis
    fig, ax = plt.subplots()
    
    # Scatter plot of the data points
    ax.scatter(my_x, my_y, label='Data Points', color='blue')
    
    # Get the limits of the x-axis
    x_min, x_max = ax.get_xlim()
    
    # Calculate y-values for the regression line
    # Ensure my_m and my_b are detached if they are PyTorch tensors
    
    my_m = my_m.detach().numpy()
    my_b = my_b.detach().numpy()
    
    y_min = my_m * x_min + my_b
    y_max = my_m * x_max + my_b
    
    # Set x-axis limits
    ax.set_xlim([x_min, x_max])
    
    # Plot the regression line
    ax.plot([x_min, x_max], [y_min, y_max], color='red', label='Regression Line')
    
    # Add labels and legend
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_title('Regression Plot')
    ax.legend()
    
    # Show the plot
    plt.show()
```

```python
regression_plot(x, y, m, b)
```

```python
def regression(my_x, my_m, my_b):
    return my_x* my_m + my_b
```

yhat is the estimate of the values of y

**Steps to calculate the correct m and b**

Step 1 : Forward Pass


Step 2 : Compare yhat with true y to calculate cost c


Step 3 : C = $g(f(x, m, b), y)$ Use chain rule to calculate gradient of C w.r.t paramaters


Step 4: adjust $m$ and  $b$ to reduce $c$ with gradient descent

```python
yhat = regression(x, m , b) # Step1 : Forward Pass -> Feeding the values of m and b to calculate yhat
```

```python
# Defining C to calculate the C . In this MSE will be since we're dealing with a regression problem
def mse(my_yhat, my_y):
    sigma = torch.sum((my_yhat - my_y)**2)
    return sigma/len(my_y)
```

```python
C = mse(yhat, y)
```

```python
C.backward() # Step 3: Using autodiff to calculate gradient of C w.r.t paramaters
```

```python
m.grad
```

```python
b.grad
```

```python
optimizer = torch.optim.SGD([m, b], lr = 0.01) # Using Gradient Descent to adjust m and b
```

```python
optimizer.step()
```

```python
# Confirm Paramters have been adjusted sesnibly
m
```

```python
b
```

```python
regression_plot(x, y, m, b)
```

```python
epochs = 1000
for epoch in range(epochs):
    
    optimizer.zero_grad()  #reset gradients to zero ; else they accumulate
    
    yhat = regression(x, m , b)  # step 1
    
    C = mse(yhat, y)   # step 2
    
    C.backward()  # step 3
    
    optimizer.step() # step 4
    
    print('Epoch {}, cost {}, m grad {}, b grad {}'.format(epoch, '% 3g' %C.item(), '% 3g' %m.grad.item(),  '% 3g' %b.grad.item() ))
```
