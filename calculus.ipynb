# Calculus of Infinitesimals

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 1000)
```

If $y = x^2 + 2x + 2$

```python
y = x**2 + 2*x + 2
```

```python
fig, ax = plt.subplots()
_ = ax.plot(x, y)
```

* There are no straight lines on the curve
* If we zoom infinitely close, however we observe curves that approach lines
* This eneables us to find a slope $m$ (tangent) anywhere on the curve 

```python
fig, ax = plt.subplots()
ax.set_xlim([-2, -0])
ax.set_ylim([0, 5])
_ = ax.plot(x, y)
```

```python
fig, ax = plt.subplots()
ax.set_xlim([-.5, 0])
ax.set_ylim([0, 5])
_ = ax.plot(x, y)
```

```python
fig, ax = plt.subplots()
ax.set_xlim([-.0005
             , 0])
ax.set_ylim([1.5, 2.5])
_ = ax.plot(x, y)
```

* As the limits approach a specific point, the curve begins to resemble a straight line, which helps in determining the slopes of the curves.

# Limits

* These can be easily calculated for continuous functions by simply substituting the limit value into the function to find the limit.

```python
fig, ax = plt.subplots()
plt.axvline(x = 0, color = 'lightgray')
plt.axhline(y = 0, color = 'lightgray')
plt.xlim(-5, 10)
plt.ylim(-10, 50)
_ = ax.plot(x, y)
```

when x is approaching 5 the value of the $y = x^2 + 2x + 5$ is equal to 25 + 10 + 2 . The plot changes to the below plot when add the limit to the plot

```python
fig, ax = plt.subplots()
plt.axvline(x = 0, color = 'lightgray')
plt.axhline(y = 0, color = 'lightgray')
plt.axvline(x = 5, color = 'gray')
plt.axhline(y = 37, color = 'gray')
plt.xlim(-5, 10)
plt.ylim(-10, 50)
_ = ax.plot(x, y)
```

For Discontinous functions

```python
def f_n (x):
    
    y = (x**2 -1)/(x-1)
    return y
```

```python
f_n(2)
```

when the value of x = 1 there is a gap in the function as the denominator is reaching zero thus resulting in division by zero error

```python
try:
    f_n(1)
except ZeroDivisionError as e : 
    print('error :', e)
```

```python
f_n(1.0000001)
```

```python
f_n(0.9999991)
```

As the value of x is reaching 1 the value of the function is approximately equal to 2 thus the limit of the function y when x approching 1 is 2

```python
y = f_n(x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x = 0, color = 'lightgray')
plt.axhline(y = 0, color = 'lightgray')
plt.xlim(-2, 5)
plt.ylim(-1, 5)
plt.axvline(x = 1, color = 'purple', linestyle = '--')
plt.axhline(y = 2, color = 'purple', linestyle = '--')
_ = ax.plot(x, y)
```
